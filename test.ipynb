```python
import psycopg2
import requests
from bs4 import BeautifulSoup
import time
import random

# Database connection
conn = psycopg2.connect(
    dbname="jobs",
    user="postgres",
    password="postgres",
    host="10.100.21.125",
    port="5432"
)
cursor = conn.cursor()

# Fetch URLs from the database
cursor.execute("SELECT id, url from raw where source = 'topcv.vn' order by id")
rows = cursor.fetchall()

def fetch_page_with_retry(url, backoff_factor=1.0):
    retries = 0
    while True:
        response = requests.get(url)
        if response.status_code == 200:
            return response
        elif response.status_code == 429:
            wait_time = backoff_factor * (2 ** retries) + random.uniform(0, 1)
            print(f"429 Too Many Requests: Retrying in {wait_time:.2f} seconds...")
            time.sleep(wait_time)
            retries += 1
        else:
            response.raise_for_status()
    return None

def update_content_and_requirement():
    for row in rows:
        job_id, url = row
        print(f"Processing job ID: {job_id}")

        response = fetch_page_with_retry(url)
        if response is None:
            print(f"Failed to fetch {url} after retries.")
            continue

        soup = BeautifulSoup(response.content, 'html.parser')

        # Extract content
        content_div = soup.select_one('#box-job-information-detail > div.job-detail__information-detail--content > div > div:nth-child(1) > div')
        content = []
        if content_div:
            content += [p.get_text(strip=True) for p in content_div.find_all('p')]
            content += [li.get_text(strip=True) for ul in content_div.find_all('ul') for li in ul.find_all('li')]

        # Extract requirement
        requirement_div = soup.select_one('#box-job-information-detail > div.job-detail__information-detail--content > div.job-description > div:nth-child(2) > div')
        requirement = []
        if requirement_div:
            requirement += [p.get_text(strip=True) for p in requirement_div.find_all('p')]
            requirement += [li.get_text(strip=True) for ul in requirement_div.find_all('ul') for li in ul.find_all('li')]

        print(f"Content: {content}")
        print(f"Requirement: {requirement}")
        # Update the database
        cursor.execute("""
            UPDATE raw
            SET content = %s, requirement = %s
            WHERE id = %s
        """, (content, requirement, job_id))

    conn.commit()

# Run the update function
update_content_and_requirement()

# Close the database connection
cursor.close()
conn.close()
```

```python
import spacy
import psycopg2
from pgvector.psycopg2 import register_vector
import numpy as np
import re
import time
import random

# Load the Vietnamese SpaCy model
nlp = spacy.load("vi_core_news_lg")

# Database connection
conn = psycopg2.connect(
    dbname="jobs",
    user="postgres",
    password="postgres",
    host="10.100.21.125",
    port="5432"
)
cursor = conn.cursor()

# Register pgvector
register_vector(cursor)

# Ensure the pgvector extension is enabled
cursor.execute("CREATE EXTENSION IF NOT EXISTS vector;")
conn.commit()

# Function to clean text (retain words and numbers only in Vietnamese)
def clean_text(text):
    text = re.sub(r'[^A-Za-zÀ-ỹ0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text)
    return text.lower().strip()

# Function to get the embedding vector for a given text
def get_embedding(text):
    doc = nlp(text)
    return doc.vector

# Function to compute Jaccard similarity
def jaccard_similarity(list1, list2):
    set1 = set(list1)
    set2 = set(list2)
    intersection = len(set1.intersection(set2))
    union = len(set1.union(set2))
    return intersection / union

# Function to fetch top 3 similar records based on content embeddings
def fetch_top_similar_embeddings(content_vector):
    cursor.execute("""
        SELECT id, global1_id, content_vector
        FROM embeddings1
        ORDER BY content_vector <-> %s LIMIT 3
    """, (content_vector,))
    content_similar = cursor.fetchall()

    return content_similar

# Get latest raw_id from the last line of file
latest_raw_id = 3000
# with open("/khanh/airflow/latest_raw_id.txt", "r") as file:
#     lines = file.readlines()
#     if lines:
#         latest_raw_id = int(lines[-1].strip())

# Fetch new records from the raw table
cursor.execute("SELECT id, title, content FROM raw WHERE source = 'topcv.vn' AND id > %s ORDER BY id", (latest_raw_id,))
new_records = cursor.fetchall()

for new_record in new_records:
    raw_id, title, content = new_record
    print(f"Processing raw ID: {raw_id}")

    # Clean and embed the content
    cleaned_title = clean_text(title)
    cleaned_content = clean_text(content)
    if not cleaned_title or not cleaned_content:
        print("title: ", cleaned_title)
        print("content: ", cleaned_content)
        print(f"Skipping raw ID: {raw_id}")
        continue

    content_vector = get_embedding(cleaned_content)

    # Fetch top 3 similar records
    content_similar = fetch_top_similar_embeddings(content_vector)

    # Check for similarity
    is_duplicate = False
    for content_sim in content_similar:
        _, global1_id, content_vec = content_sim
        print(f'raw_id: {raw_id}, global1_id: {global1_id}')
        content_sim_value = jaccard_similarity(cleaned_content.split(), clean_text(' '.join(map(str, content_vec))).split())
        if content_sim_value > 0.8:
            is_duplicate = True
            print(f"Duplicate found with global ID: {global1_id} and raw ID: {raw_id}")
            # Store to file
            with open("/khanh/airflow/duplicates.txt", "a") as file:
                file.write(f"{raw_id}, {global1_id}\n")
            break

    # Insert into global and embeddings tables if not a duplicate
    if not is_duplicate:
        cursor.execute("""
            INSERT INTO global1 (title, content, source)
            VALUES (%s, %s, 'topcv.vn') RETURNING id
        """, (title, content))
        global1_id = cursor.fetchone()[0]

        cursor.execute("""
            INSERT INTO embeddings1 (global1_id, content_vector)
            VALUES (%s, %s)
        """, (global1_id, content_vector))

# Save latest raw_id to file
cursor.execute("SELECT max(id) FROM raw")
latest_raw_id = cursor.fetchone()[0]
with open("/khanh/airflow/latest_raw_id.txt", "a") as file:
    file.write('\n' + str(latest_raw_id))

conn.commit()
cursor.close()
conn.close()
```
